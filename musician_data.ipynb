```python
import re
import pandas as pd
import requests

def dbr_clean(string_list):
    clean = ['dbr:', ' \t']
    new_name = re.split(",", re.sub(clean[1], ',', re.sub(clean[0], '', string_list, count=0, flags=0), count=0, flags=0))
    return new_name

def db_clean(string_list, token):
    new_string = re.sub(token, '', string_list, count=0, flags=0)
    new_string = re.sub(' \t', ',', new_string, count=0, flags=0)
    return new_string

def ch_clean(string_list):
    new_string = re.sub(r'[^\x00-\x7F]+', '', string_list, count=0, flags=0)
    new_string = re.sub('?', '', new_string, count=0, flags=0)
    return new_string

def union(list1, list2):
    final_list = list(set(list1) | set(list2))
    return final_list


def collect_namelist(data):
    data_list=[]
    for v in range(len(data)):
        if data['Property'][v]=='is dbo:associatedBand  of':
            data_list = union(data_list, dbr_clean(data['Value'][v]))
        elif data['Property'][v]=='dbo:associatedBand':
            data_list = union(data_list, dbr_clean(data['Value'][v]))
    return data_list


def update_list(name_list):
    for name in name_list:
        url = 'https://dbpedia.org/page/' + name
        try:
            html = requests.get(url)
            res_data = pd.read_html(url)
            data = res_data[0]
        except Exception as e:
            print(name)
            name_list.remove(name)
        if len(data)<=10:
            print('data Not Found')
            print(name)
            name_list.remove(name)
    return name_list


def find_musician_name(name_list):
    new_list=[]
    for name in name_list:
        print(name)
        url = 'https://dbpedia.org/page/' + name
        html = requests.get(url)
        res_data = pd.read_html(url)
        data = res_data[0]
        print(collect_namelist(data))
        new_list = union(new_list, collect_namelist(data))
    return union(new_list, name_list)
```

```python
def find_musician_data(mu_name_list):
    variable = ['yearsActive', 'activeYearsStartYear', 'activeYearsEndYear']
    data_matrix = pd.DataFrame(index=mu_name_list, columns=variable)
    for N in range(len(mu_name_list)):
        print(mu_name_list[N])
        url = 'https://dbpedia.org/page/' + mu_name_list[N]
        html = requests.get(url)
        res_data = pd.read_html(url)
        data = res_data[0]
        col = ['Property', 'Value']
        for i1 in col:
            for i2 in range(len(data)):
                clean_token = ['dbo:', 'dbr:', 'dbp:', 'dct:']
                for token in clean_token:
                    data[i1][i2] = db_clean(data[i1][i2], token)
        variable = ['yearsActive', 'activeYearsStartYear', 'activeYearsEndYear']
        for j1 in variable:
            for j2 in range(len(data)):
                if data['Property'][j2] == j1:
                    data_matrix[j1][N] = data['Value'][j2]
    return data_matrix
```
